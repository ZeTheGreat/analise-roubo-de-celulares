```python
import pandas as pd
from pandas import Series, DataFrame
import folium
from folium import plugins
import numpy as np
```

```python
path = 'dataset/'
filename = 'DadosBO_2018_9(FURTO DE VEÍCULOS)'
extension = '.xls'
```

```python
def dataFrame():
    dfList = list()
    
    df= pd.DataFrame(pd.read_csv(path+filename+extension, sep='\\t', header=None, engine='python'))
    dfList.append(df)

    df = pd.concat(dfList)
    return df
```

```python
df = dataFrame()
```

```python
df.reset_index()
```

```python
# selecionando as colunas de geoloocalização que serão utilizadas
localidades = df[[17, 18]]
```

```python
# verificando se há registros nulos
localidades.isnull().sum()
```

```python
# deletando/trocando por 0 registros nulos
localidades.dropna(inplace=True)
localidades.fillna(0)
```

```python
# verificando se ainda há registros nulos
localidades.isnull().sum()
```

```python
# função para limpar e castear para float as coordenadas provindas do DataFrame que inicialmente eram str.
def strToFloat(lista):
    tirar = "\x00"
    aux = []
    for i in lista:
        for j in range(0, len(tirar)):
            if type(i) == str:
                i = i.replace(tirar[j], "")
                i = i.replace(",", ".")
                if bool(i):
                    aux.append(float(i))
    return aux
```

```python
locais = []
old_latitude = localidades[17].values
latitude = np.delete(old_latitude, 0)
old_longitude = localidades[18].values
longitude = np.delete(old_longitude, 0)

latitude = strToFloat(latitude)
longitude = strToFloat(longitude)
    
```

```python
mapa = folium.Map(location=[-23.1791, -45.8872],tiles='OpenStreetMap',zoom_start=12)

for la,lo in zip(latitude,longitude):
    locais.append([la, lo])
    
mapa.add_child(plugins.HeatMap(locais))
```
