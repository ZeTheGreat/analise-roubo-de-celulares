```python
path = 'dataset/'
filename = 'Roubo de Celular - Setembro'
extension = '.csv'
```

```python
def populaDataFrame():
    dfList = list()
    
    df= pd.DataFrame(pd.read_csv(path+filename+extension, sep='\\t', header=None, engine='python'))
    dfList.append(df)

    df = pd.concat(dfList)
    return df
    
```
